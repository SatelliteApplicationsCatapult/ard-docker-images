```python
esa_s2ids = list(esa_l1c.title.values)
esa_s2ids[0]
```

```python
sedas_prods = []
unavailable = []
for s2id in tqdm.tqdm(esa_s2ids[:]):
    singleProduct = sedas.search_product(s2id)
    try:
        sedas_prods.append(singleProduct[0])
    except:
        unavailable.append(s2id)
sedas_prods = pd.DataFrame(sedas_prods)
```
